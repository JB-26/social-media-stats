# USA Data
This notebook is about exploring USA data

```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
```

```python
us_df = pd.read_csv('social_media-US-monthly-201305-202305.csv')
```

```python
us_df.head()
```

```python
# important to convert the date column to datetime!
us_df['Date'] = pd.to_datetime(us_df['Date'])
```

```python
us_df.info()
```

```python
p = figure(width=800, height=800, title="USA Data",  x_axis_type = 'datetime', x_axis_label = 'Date', y_axis_label= 'Percent')
p.line(y=us_df['Facebook'], x=us_df['Date'], line_width=2, color='red', legend_label='Facebook')
p.line(y=us_df['Twitter'], x=us_df['Date'], line_width=2, color='green', legend_label='Twitter')
p.line(y=us_df['Pinterest'], x=us_df['Date'], line_width=2, color='grey', legend_label='Pinterest')
p.line(y=us_df['Tumblr'], x=us_df['Date'], line_width=2, color='black', legend_label='Tumblr')
p.line(y=us_df['Instagram'], x=us_df['Date'], line_width=2, color='blue', legend_label='Instagram')
p.line(y=us_df['reddit'], x=us_df['Date'], line_width=2, color='purple', legend_label='reddit')
p.line(y=us_df['YouTube'], x=us_df['Date'], line_width=2, color='cyan', legend_label='YouTube')
show(p)
```
