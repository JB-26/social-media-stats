# Japan Data
This notebook is about exploring Japan data

```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
```

```python
jp_df = pd.read_csv("social_media-JP-monthly-201305-202305.csv")
```

```python
jp_df.head()
```

```python
# important to convert the date column to datetime!
jp_df['Date'] = pd.to_datetime(jp_df['Date'])
```

```python
jp_df.info()
```

```python
p = figure(width=800, height=800, title="USA Data",  x_axis_type = 'datetime', x_axis_label = 'Date', y_axis_label= 'Percent')
p.line(y=jp_df['Facebook'], x=jp_df['Date'], line_width=2, color='red', legend_label='Facebook')
p.line(y=jp_df['Twitter'], x=jp_df['Date'], line_width=2, color='green', legend_label='Twitter')
p.line(y=jp_df['Pinterest'], x=jp_df['Date'], line_width=2, color='grey', legend_label='Pinterest')
p.line(y=jp_df['Tumblr'], x=jp_df['Date'], line_width=2, color='black', legend_label='Tumblr')
p.line(y=jp_df['Instagram'], x=jp_df['Date'], line_width=2, color='blue', legend_label='Instagram')
p.line(y=jp_df['reddit'], x=jp_df['Date'], line_width=2, color='purple', legend_label='reddit')
p.line(y=jp_df['YouTube'], x=jp_df['Date'], line_width=2, color='cyan', legend_label='YouTube')
show(p)
```
